```python
import psycopg2
from faker import Faker  # Librería para generar datos falsos

# Parámetros de conexión
dbname = "NotesStudy"
user = "postgres"
password = "1234"
host = "localhost"
port = "5432"

# Crear una instancia de Faker
faker = Faker()

# Generar 100 datos de prueba para cada tabla
alumnos = [(i, faker.pyfloat(min_value=0, max_value=5), faker.random_int(min=1, max=8),
            faker.random_int(min=2000, max=2022), faker.random_element(["Masculino", "Femenino"]), 
            faker.random_int(min=16, max=26)) for i in range(1, 101)]

apoyos = [(i, faker.city(), faker.random_int(min=1, max=6), faker.random_element(["Bus", "Metro", "Bicicleta"]), 
           faker.random_int(min=1, max=6), faker.random_int(min=0,max=10000000),
           faker.random_int(min=0, max=2), faker.random_int(min=5, max=120)) for i in range(1, 101)]

asistencias = [(i, faker.random_int(min=0, max=20), faker.random_int(min=0, max=20)) for i in range(1, 101)]

# Generar horarios en el formato de intervalo
horarios = [f"{faker.random_int(min=1, max=12)}:{faker.random_element(['00', '15', '30', '45'])}{faker.random_element(['am', 'pm'])} - {faker.random_int(min=1, max=12)}:{faker.random_element(['00', '15', '30', '45'])}{faker.random_element(['am', 'pm'])}" for _ in range(1, 101)]

# Insertar datos en la tabla materia
materias = [(i, faker.catch_phrase(), faker.random_int(min=1, max=5),
             horarios[i-1], faker.random_int(min=1, max=100)) for i in range(1, 101)]


profesores = [(i, faker.name(), faker.random_int(min=1, max=8), faker.random_element(["A", "B", "C"]),
               faker.random_int(min=1, max=50), faker.job(), faker.job(),
               faker.random_int(min=1, max=100)) for i in range(1, 101)]

notas = [(i, j, faker.pyfloat(min_value=0, max_value=5),
          faker.random_element(["A", "B", "C"]), faker.date_time_this_year(),
          faker.random_int(min=1, max=10), faker.random_int(min=1, max=10)) for i in range(1, 101) for j in range(1, 101)]

# Insertar datos de prueba en la base de datos
try:
    connection = psycopg2.connect(
        dbname=dbname, user=user, password=password, host=host, port=port
    )
    print("Conexión exitosa a la base de datos")

    cursor = connection.cursor()

    # Insertar datos en la tabla alumno
    insert_query = "INSERT INTO alumno (id_alumno, promedio, semestre, anio_entrada, sexo, edad) VALUES (%s, %s, %s, %s, %s, %s)"
    cursor.executemany(insert_query, alumnos)

    # Insertar datos en la tabla apoyoeconomico
    insert_query = "INSERT INTO apoyoeconomico (id_alumno, localidad, estrato, transporte, personas_en_casa, apoyo_universitario, num_hijos, tiempo_transporte) VALUES (%s, %s, %s, %s, %s, %s, %s, %s)"
    cursor.executemany(insert_query, apoyos)

    # Insertar datos en la tabla asistencia
    insert_query = "INSERT INTO asistencia (id_alumno, faltas_justificadas, faltas_injustificadas) VALUES (%s, %s, %s)"
    cursor.executemany(insert_query, asistencias)

    # Insertar datos en la tabla profesor
    insert_query = "INSERT INTO profesor (id_profesor, nombre_profesor, semestre, grupo, num_estudiantes, facultad, titulo_universitario, id_materia) VALUES (%s, %s, %s, %s, %s, %s, %s, %s)"
    cursor.executemany(insert_query, profesores)

    # Insertar datos en la tabla materia
    insert_query = "INSERT INTO materia (id_materia, nombre_materia, creditos, horario, profesor) VALUES (%s, %s, %s, %s, %s)"
    cursor.executemany(insert_query, materias)


    # Insertar datos en la tabla nota
    insert_query = "INSERT INTO nota (id_materia, id_alumno, nota_final, grupo, horario, registro_asistencia, horas_estudiadas) VALUES (%s, %s, %s, %s, %s, %s, %s)"
    cursor.executemany(insert_query, notas)

    connection.commit()
    print("Datos de prueba insertados correctamente")

    cursor.close()

except psycopg2.Error as e:
    print("Error al conectarse a la base de datos:", e)
```
